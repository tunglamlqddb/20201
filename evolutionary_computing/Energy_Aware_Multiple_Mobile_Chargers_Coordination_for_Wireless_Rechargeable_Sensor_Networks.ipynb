```python
def find_sigma_in_cycle(E: np.array, n: int):    # E la mang co M phan tu

  status = [1] * M   # check dead or alive of MC

  min_num_of_rounds = (E / max_energy_and_time).astype(int)
  #min_num_of_rounds = np.array([4,4,4,4,1,2])    # test thử

  res = []
  max_rounds_per_cycle = 0
  num_of_avail_MCs = 0

  while True:
    max_rounds_per_cycle += 1
    num_of_avail_MCs += sum(status)
    res.append(num_of_avail_MCs)
    if num_of_avail_MCs >= n:
      return max_rounds_per_cycle, res
    else: 
      min_num_of_rounds -= 1
      for i in range(M):
        if min_num_of_rounds[i] == 0:
          status[i] = 0
        elif min_num_of_rounds[i] < 0:
          if min_num_of_rounds[i] + PHI == 0:
            status[i] = 1
            min_num_of_rounds[i] = (E[i] / max_energy_and_time).astype(int)
```

```python
def find_index_of_avai_MCs(E_l):      # in round L, E_l gom M phan tu
  res = []
  for i in range(M):
    if E_l[i] >= max_energy_and_time + D_MAX * e:
      res.append(i)
      round_min_num_of_rounds[i] = 608  
    else:
      if round_min_num_of_rounds[i] == 608:   # this MS needs to return to BS RIGHT AT THIS ROUND
        round_min_num_of_rounds[i] = 0
  return res
```

```python
def update_after_MPSP(list_avai_MCs, E_l, consumed, E_after_charging):       # hàm này được chạy trước hàm find_index_of_avai_MCs, để đảm đã bao gồm các MC có thể quay trở lại từ BS
  E_l_plus_1 = [0] * M
  for j in list_avai_MCs:
    E_l_plus_1[j] = E_l[j] - consumed[j]

  list_not_avai_MCs = set(list(range(M))).difference(set(list_avai_MCs))
  for j in list_not_avai_MCs:
    round_min_num_of_rounds[j] -= 1
    if round_min_num_of_rounds[j] + PHI == 0:
      E_l_plus_1[j] = E_after_charging           # cập nhật những MC có thể quay trở lại từ BS
    else:
      E_l_plus_1[j] = E_l[j]
  
  return E_l_plus_1
```

```python
def calculate_total_time_for_round_L(q, t, g, h_k_l, p_k_l):  
  res = 0
  for i in range(p_k_l):
    for j in range(h_k_l):
      if q[i][j].solution_value() > 0:
       # print("Node {}, time {} + {} ".format(new_index_node[i], t[i][j].solution_value(), g[i][j].solution_value()))
        res = max(res, t[i][j].solution_value() + g[i][j].solution_value())
  return res

def get_position_of_MCs_in_round_L(q,t,g,h_k_l,p_k_l, new_index_MC, new_index_node):
  position = defaultdict(int)
  for j in range(h_k_l):
    for i in range(p_k_l):
      if q[i][j].solution_value() > 0:
        position[new_index_MC[j]] = new_index_node[i]
  return position

def update_e_k(q,t,g,h_k_l,p_k_l, new_index_node):
  total_time_for_round_L = calculate_total_time_for_round_L(q,t,g,h_k_l,p_k_l)
  for i in range(len(dynamic_e_k)):
    dynamic_e_k[i] -= total_time_for_round_L * consumption_energy_rate[i]
  for j in range(h_k_l):
    for i in range(p_k_l):
      if q[i][j].solution_value() > 0:
        # cộng lại phần bên trên bị trừ
        dynamic_e_k[new_index_node[i]] += total_time_for_round_L * consumption_energy_rate[new_index_node[i]]  
        # trừ đi năng lượng đợi MC di chuyển tới
        dynamic_e_k[new_index_node[i]] -= g[i][j].solution_value() * consumption_energy_rate[new_index_node[i]] 
        # cộng thêm năng lượng được sạc 
        dynamic_e_k[new_index_node[i]] += t[i][j].solution_value() * p_r 
        # trừ đi năng lượng trong lúc đợi MC lâu nhất sạc xong
        dynamic_e_k[new_index_node[i]] -= (total_time_for_round_L - t[i][j].solution_value() - g[i][j].solution_value()) * consumption_energy_rate[new_index_node[i]]    
```

```python
from __future__ import print_function
from ortools.linear_solver import pywraplp

def solve(new_index_node, new_index_MC, prev_pos_MC, p_k_l, h_k_l, t_l_max, theta_k_l, distance, max_energy_and_time, v, E_low, E_high, p_r, e, p0):
  # E_low, E_high: index from 0->p_k_l
  # t_l_max: index is the ACTUAL index of node


  # Create the mip solver with the SCIP backend.
  solver = pywraplp.Solver.CreateSolver('SCIP')

  # Create variables
  q = [[solver.IntVar(0, 1, 'q_ij') for _ in range(h_k_l)] for _ in range(p_k_l)]
  t = [[solver.NumVar(0, t_l_max[new_index_MC[j]], 't_ij') for j in range(h_k_l)] for _ in range(p_k_l)]
  g = [[solver.NumVar(0, theta_k_l, 'g_ij') for _ in range(h_k_l)] for _ in range(p_k_l)]

  # Create constraints
  # print("Các ràng buộc:")

  #(10)
  for i in range(p_k_l):
    if E_high[i] <= E_low[i]:
      constraint1 = solver.Constraint(E_high[i], E_high[i])
    else:
      constraint1 = solver.Constraint(E_low[i], E_high[i])
    # print('\t (10): {} <= t_i_j '.format(E_low[i]), end='')
    for j in range(h_k_l):
      constraint1.SetCoefficient(t[i][j], p_r)
    #   if j == h_k_l-1:
    #     print('{}*t_{}_{}'.format(p_r, new_index_node[i],new_index_MC[j]), end=" ")
    #   else:
    #     print('{}*t_{}_{}'.format(p_r, new_index_node[i],new_index_MC[j]), end=" + ")
    # print('<= {}'.format(E_high[i]))

  #(13)
  for j in range(h_k_l):
    constraint2 = solver.Constraint(0, 1)
    for i in range(p_k_l):
      constraint2.SetCoefficient(q[i][j], 1)

  #(14)
  for i in range(p_k_l):
    constraint3 = solver.Constraint(1,1)
    for j in range(h_k_l):
      constraint3.SetCoefficient(q[i][j], 1)

  #(15)
  for i in range(p_k_l):
    for j in range(h_k_l):
      constraint4 = solver.Constraint(-t_l_max[new_index_MC[j]], 0)
      constraint4.SetCoefficient(t[i][j], 1)
      constraint4.SetCoefficient(q[i][j], -t_l_max[new_index_MC[j]])
      # print("\t (15): {} <= t_{}_{} - {}*q_{}_{} <= 0".format(-t_l_max[new_index_MC[j]], new_index_node[i],new_index_MC[j], t_l_max[new_index_MC[j]], new_index_node[i],new_index_MC[j]))
  

  #(16)
  for i in range(p_k_l):
    for j in range(h_k_l):
      constraint5 = solver.Constraint(-max_energy_and_time, 0)
      constraint5.SetCoefficient(q[i][j], distance[new_index_node[i]][prev_pos_MC[j]] / v)
      constraint5.SetCoefficient(g[i][j], -1)
      # print("\t (16): -{} <= {}*q_{}_{} - g_{}_{} <= 0".format(max_energy_and_time, distance[new_index_node[i]][prev_pos_MC[j]] / v, new_index_node[i],new_index_MC[j], new_index_node[i],new_index_MC[j]))

  #(17)
  for i in range(p_k_l):
    for j in range(h_k_l):
      constraint6 = solver.Constraint(0, theta_k_l)
      constraint6.SetCoefficient(t[i][j], 1)
      constraint6.SetCoefficient(g[i][j], 1)
      # print('\t (17): 0 <= t_{}_{} + g_{}_{} <= {}'.format(new_index_node[i],new_index_MC[j],new_index_node[i],new_index_MC[j],theta_k_l), end='\n')
      

  # Create objective function
  obj = solver.Objective()
  for i in range(p_k_l):
    for j in range(h_k_l):
      obj.SetCoefficient(q[i][j], distance[new_index_node[i]][prev_pos_MC[j]] * e)
      obj.SetCoefficient(t[i][j], p0)
  obj.SetMinimization()

  # Solver

  status = solver.Solve()
  # Calculate consumed energy of each avai MC in this round
  if status != pywraplp.Solver.OPTIMAL:
    print("The problem does not have an optimal solution!")
  consumed = defaultdict(int)
  print('Objective value =', obj.Value())
  for i in range(p_k_l):
    for j in range(h_k_l):
      if q[i][j].solution_value() > 0:
        print("\tMC {} charged node {}".format(new_index_MC[j], new_index_node[i]), end=': ')
        print("Node {}, time {} + {} ".format(new_index_node[i], t[i][j].solution_value(), g[i][j].solution_value()))
        consumed[new_index_MC[j]] = t[i][j].solution_value() * p0 + distance[new_index_node[i]][prev_pos_MC[j]] * e
  # get position of avai MCs in this round
  position = get_position_of_MCs_in_round_L(q,t,g,h_k_l,p_k_l, new_index_MC,new_index_node)

  # update total time before
  time_for_this_round = calculate_total_time_for_round_L(q,t,g,h_k_l,p_k_l)
  # tính lại e_k sau mỗi round
  update_e_k(q,t,g, h_k_l, p_k_l, new_index_node)


  return position, time_for_this_round, consumed, obj.Value()
```
